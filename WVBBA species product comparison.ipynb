We investigated the ability of the 2nd WV Breeding Bird Atlas (2011-2015) to model and detect species.

## Methods
We combined lists of species that USGAP and the WVBBA produced various data products (maps) for.  WVBBA wasn't able to model all species that they detected.  Additionally, they were able to model the distribution of occurrence for some species for which density couldn't be modeled, and evidence of breeding was not obtained for every species that was detected.

```python
import pandas as pd
import repo_functions as fun 
import numpy as np
pd.set_option('display.width', 2000)
pd.set_option('display.max_colwidth', 400)
pd.set_option('display.max_rows', 400)
pd.set_option('display.max_columns', 15)

# Read in the spp crosswalk
df = pd.read_csv(fun.dataDir + "SpeciesLists/WV_GAP_Atlas3.csv", header=0, 
                        names=["common_name", "GAP_sci_name", "WVBBA_sci_name", "GAP_code", "WV_code", 'GAP_modeled',
                               'detected', 'occupancy_map', 'density_map', 'change_map', 'breeding_evid_map', 
                               "GAP_habitat(ha)", "point_count_individuals", "survey_points", "count_rate", "%_of_blocks", 
                               "notes"],
                       dtype={"common_name":str, "WVBBA_sci_name":str, "%_of_blocks":float, "WV_code":str, })

# Drop the GAP columns, they aren't used in this analysis
df = df.drop(["GAP_sci_name", "GAP_code", "GAP_modeled", "GAP_habitat(ha)", "notes"], axis=1)

print("The first 5 records")
df.head().T
```

### *Exceptions*
Some species entries have been identified as problematic, so we exluded them from this summary.

__*Solitary Vireo*__ -- this concept is equivalent to Blue-headed Vireo, which is also in the tables.

__*Brewster's Warbler*__ and __*Lawrence's Warbler*__ -- these hybrids were not modeled by GAP.

__*Slate-colored Junco*__ -- equivalent to Dark-eyed Junco, which is also in the tables.

```python
drop = ["Solitary Vireo", "Brewster's Warbler", "Lawrence's Warbler", "Slate-colored Junco"]
df = df[df["common_name"].isin(drop) == False]
```

## Results

### Detected, but no evidence of breeding

```python
nobreed = (df[(df['detected'] == 1) & 
        (df['breeding_evid_map'] == 0)]
     .drop(["WVBBA_sci_name", "WV_code"], axis=1)
     .set_index(["common_name"]))
nobreed
```

### Occupancy mapped, but not density

```python
nodensity = (df[(df['occupancy_map'] == 1) & 
        (df['density_map'] == 0)]
     .drop(["WVBBA_sci_name", "WV_code"], axis=1)
     .set_index(["common_name"]))
nodensity
```

### Breeding evidence mapped, but not change

```python
nochange = (df[(df['breeding_evid_map'] == 1) & 
        (df['change_map'] == 0)]
     .drop(["WVBBA_sci_name", "WV_code"], axis=1)
     .set_index(["common_name"]))
nochange
```

### Density mapped, but not change in breeding

```python
nobreed = (df[(df['density_map'] == 1) & 
        (df['change_map'] == 0)]
     .drop(["WVBBA_sci_name", "WV_code"], axis=1)
     .set_index(["common_name"]))
nobreed
```

### Occupancy mapped, but not change in breeding

```python
noch = (df[(df['occupancy_map'] == 1) & 
        (df['change_map'] == 0)]
     .drop(["WVBBA_sci_name", "WV_code"], axis=1)
     .set_index(["common_name"]))
noch
```

### Density mapped, but not occupancy

```python
noocc = (df[(df['occupancy_map'] == 0) & 
        (df['density_map'] == 1)]
     .drop(["WVBBA_sci_name", "WV_code"], axis=1)
     .set_index(["common_name"]))
noocc
```
